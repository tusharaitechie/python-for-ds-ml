### Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

var = sns.load_dataset("penguins")
var.head(5)
```

```python
sns.scatterplot(x="bill_length_mm" , y="bill_depth_mm", data= var)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm" , y="bill_depth_mm", data= var, hue="sex")
plt.show()
```

```python
var = sns.load_dataset("penguins").head(20)
var
```

```python
sns.scatterplot(x="bill_length_mm" , y="bill_depth_mm", data= var, hue="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm" , y="bill_depth_mm", data= var, hue="sex", style="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm" , y="bill_depth_mm", data= var, hue="sex", style="sex", size="sex")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm" , y="bill_depth_mm", data= var, hue="sex", style="sex", sizes=(100, 40))
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm" , y="bill_depth_mm", data= var, hue="sex", style="sex", 
                sizes=(100, 40), palette="gist_ncar", alpha=1)
plt.show()
```

```python
m={"Male": "*", "Female": "o"}
sns.scatterplot(x="bill_length_mm" , y="bill_depth_mm", data= var, hue="sex", style="sex", 
                sizes=(100, 40), palette="gist_ncar", alpha=1, markers=m)
plt.show()
```
