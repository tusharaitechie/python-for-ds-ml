### Violin Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var.head(5)
```

```python
sns.violinplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="time")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, linewidth=2)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, linewidth=2, palette="Dark2", hue="day", legend=False)
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=var, order=["Dinner", "Lunch"])
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, linewidth=2, palette="Dark2", hue="day", legend=False, saturation=0.4)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, linewidth=2, color="y")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="sex", split=True)
plt.show()
```

```python
sns.violinplot(x="total_bill", y="day", data=var, hue="sex")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=var, order=["Dinner", "Lunch"], inner="quart")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=var, order=["Dinner", "Lunch"], inner="box")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=var, order=["Dinner", "Lunch"], inner="point")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=var, order=["Dinner", "Lunch"], inner="stick")
plt.show()
```

```python
sns.violinplot(x=var["total_bill"])
plt.show()
```

```python
sns.violinplot(y=var["total_bill"])
plt.show()
```
