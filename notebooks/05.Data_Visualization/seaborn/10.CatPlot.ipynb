```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the "tips" dataset
var = sns.load_dataset("tips").head(20)

var.head(5)
```

```python
# Create the plot using sns.catplot()
sns.catplot(x="size", y="tip", data=var)

# Show the plot
plt.show()
```

```python
sns.catplot(x="size", y="tip", data=var, hue="sex")

plt.show()
```

```python
sns.catplot(x="size", y="tip", data=var, hue="sex", palette="Oranges")

plt.show()
```

```python
sns.catplot(x="size", y="tip", data=var, hue="sex", palette="Oranges", height=3)

plt.show()
```

```python
sns.catplot(x="size", y="tip", data=var, hue="sex", palette="Oranges", kind="bar")

plt.show()
```

```python
sns.catplot(x="size", y="tip", data=var, hue="sex", palette="Oranges", kind="point")

plt.show()
```

```python
sns.catplot(x="size", y="tip", data=var, hue="sex", palette="Oranges", kind="boxen")

plt.show()
```
