```python
sns.heatmap(data=var)
```

```python
var1 = np.linspace(1, 10, 10).reshape(2, 5)
var1
```

```python
sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=True)
```

```python
ar = np.array([["a0", "a1", "a2", "a3", "a4"], 
               ["b0", "b1", "b2", "b3", "b4"]])
ar
```

```python
sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=ar, fmt="s")
plt.show()
```

```python
y={"fontsize":10, "color":"r"}
sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=True, annot_kws=y)
```

```python
y={"fontsize":20, "color":"r"}
sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=True, annot_kws=y)
```

```python
y={"fontsize":20, "color":"r"}
sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=True, annot_kws=y, linewidth=4)
```

```python
y={"fontsize":20, "color":"r"}
sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=True, annot_kws=y, linewidth=4, linecolor="y")
```

```python
y={"fontsize":20, "color":"r"}
sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=True, annot_kws=y, linewidth=4, linecolor="y", cbar=False)
```

```python
y={"fontsize":20, "color":"r"}
sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=True, annot_kws=y, linewidth=4, linecolor="y", cbar=False, xticklabels=False)
```

```python
y={"fontsize":20, "color":"r"}
sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=True, annot_kws=y, linewidth=4, linecolor="y", cbar=False, xticklabels=False, yticklabels=False)
```

```python
y={"fontsize":20, "color":"r"}
sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=True, annot_kws=y, linewidth=4, 
            linecolor="y", cbar=False, xticklabels=False, yticklabels=False)
plt.ylabel("exercise")
plt.show()
```

```python
y={"fontsize":20, "color":"r"}
sns.heatmap(var1, vmin=0, vmax=10, cmap='PuOr', annot=True, annot_kws=y, linewidth=4, 
            linecolor="y", cbar=False, xticklabels=False, yticklabels=False)
plt.ylabel("exercise")
#sns.set(font_scale=5)
plt.show()
```
