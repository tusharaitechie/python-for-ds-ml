### Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = sns.load_dataset("penguins")

var.head(5)
```

```python
sns.barplot(x=var.island, y=var.bill_depth_mm)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex")
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order = order_1)
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order = order_1, hue_order=["Female", "Male"])
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order = order_1, 
            hue_order=["Female", "Male"], errorbar=('ci', 10))
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order = order_1, 
            hue_order=["Female", "Male"], errorbar=('ci', 10), n_boot=2)
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order = order_1, 
            hue_order=["Female", "Male"], errorbar=('ci', 10), n_boot=2, orient="v")
plt.show()
```

```python
sns.barplot(x="bill_depth_mm", y="bill_length_mm", data=var, orient="v")
plt.show()
```

```python
sns.barplot(x="bill_depth_mm", y="bill_length_mm", data=var, orient="h")
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, color="b")
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order = order_1, 
            hue_order=["Female", "Male"], errorbar=('ci', 10), n_boot=2, orient="v", 
           palette="icefire")
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order = order_1, 
            hue_order=["Female", "Male"], errorbar=('ci', 10), n_boot=2, orient="v", 
           saturation=10)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", err_kws={'color': 'g'})
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", err_kws={'color': 'g', 'linewidth': 12})
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", err_kws={'color': 'g'}, dodge=True)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", err_kws={'color': 'g'}, dodge=False)
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order = order_1, 
            hue_order=["Female", "Male"], n_boot=2,  
           saturation=100, alpha=0.5)
plt.show()
```

```python
sns.set(style="darkgrid")
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=var, hue="sex", order = order_1, 
            hue_order=["Female", "Male"], n_boot=2,  
           saturation=100, alpha=0.5)
plt.show()
```
