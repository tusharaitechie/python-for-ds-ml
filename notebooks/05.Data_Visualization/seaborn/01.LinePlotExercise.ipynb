### Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = [1, 2, 3, 4, 5, 6, 7]
var_1 = [2, 3, 4, 1, 5, 6, 7]

x_1 = pd.DataFrame({"var": var, "var_1":var_1})

sns.lineplot(x="var", y="var_1", data=x_1) # graph using seaborn
plt.show() # this function is from matplot
```

```python
plt.plot(var, var_1) # graph using matplot
plt.show()
```

```python
# now we will use the dataset from github : https://github.com/mwaskom/seaborn-data
```

```python
y_1 = sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x="bill_length_mm", y = "flipper_length_mm", data = y_1)

plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y = "flipper_length_mm", data = y_1, hue="sex")

plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y = "flipper_length_mm", data = y_1, hue="sex", size=10)

plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y = "flipper_length_mm", data = y_1, hue="sex", style="sex")

plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y = "flipper_length_mm", data = y_1, hue="sex", style="sex", palette="Accent")

plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y = "flipper_length_mm", data = y_1, hue="sex", style="sex", palette="Accent", markers=["o", ">"])

plt.show()
```

```python
y_1 = sns.load_dataset("penguins").head(20)
y_1
```

```python
sns.lineplot(x="bill_length_mm", y = "flipper_length_mm", data = y_1, hue="sex", style="sex", palette="Accent", markers=["o", ">"], dashes=True)

plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y = "flipper_length_mm", data = y_1, hue="sex", style="sex", palette="Accent", markers=["o", ">"], dashes=False)

plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y = "flipper_length_mm", data = y_1, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"], dashes=False, legend=False)

plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y = "flipper_length_mm", data = y_1, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"], dashes=False, legend="full")

plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y = "flipper_length_mm", data = y_1, hue="sex", style="sex", 
             palette="Accent", markers=["o", ">"], dashes=False, legend="full")

plt.grid()
plt.title("python")
plt.show()
```
