### Strip Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

var = sns.load_dataset("tips")
var.head(5)                       
```

```python
sns.stripplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="rocket_r")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="rocket_r", linewidth=1)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="rocket_r", linewidth=1, edgecolor="r")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="rocket_r", linewidth=1, edgecolor="r", jitter=1)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="rocket_r", 
              linewidth=1, edgecolor="r", jitter=1, size=20)
plt.show()
```

```python
m={"male": "o", "female": "<"}
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="rocket_r", 
              linewidth=1, edgecolor="r", jitter=1, size=20, marker="<")
plt.show()
```

```python
m={"male": "o", "female": "<"}
sns.stripplot(x="day", y="total_bill", data=var, hue="sex", palette="rocket_r", 
              linewidth=1, edgecolor="r", jitter=1, size=20, marker="<", alpha=0.2)
plt.show()
```
