### Seaborn Styling Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

var = sns.load_dataset('tips')
var.head(5)
```

```python
sns.boxplot(y=var["total_bill"])
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=var)
plt.grid()
plt.show()
```

```python
sns.set_style("dark")
sns.barplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.barplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.set_style("ticks")
sns.barplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.set_style("dark")
sns.barplot(x="day", y="total_bill", data=var)
sns.despine()
plt.show()
```

```python
sns.set_style("dark")
plt.figure(figsize=(12, 3))
sns.barplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.set_style("dark")
plt.figure(figsize=(3, 12))
sns.barplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(12, 5))
sns.barplot(x="day", y="total_bill", data=var)
sns.despine()
plt.show()
```

```python
sns.set_style("white")
sns.set_context("talk", font_scale=1)
sns.barplot(x="day", y="total_bill", data=var, palette='cool')
plt.show()
```
