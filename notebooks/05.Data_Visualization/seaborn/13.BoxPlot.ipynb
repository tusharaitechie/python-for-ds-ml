### Box Plot using seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = sns.load_dataset("tips")

var.head(5)
```

```python
sns.boxplot(x= "day", y="total_bill", data=var)
plt.show()
```

```python
sns.boxplot(x= "day", y="total_bill", data=var, hue="sex")
plt.show()
```

```python
sns.boxplot(x= "day", y="total_bill", data=var, hue="sex", palette='dark:g')
plt.show()
```

```python
sns.boxplot(x= "day", y="total_bill", data=var, hue="sex", palette='dark:g', order=["Fri", "Sun", "Thur", "Sat"])
plt.show()
```

```python
sns.boxplot(x= "day", y="total_bill", data=var, hue="sex", palette='dark:g', 
            order=["Fri", "Sun", "Thur", "Sat"], showmeans=True, meanprops={"marker": "+", "markeredgecolor" : "b"})
plt.show()
```

```python
sns.boxplot(x= "day", y="total_bill", data=var, hue="sex", palette='plasma', 
            order=["Fri", "Sun", "Thur", "Sat"], showmeans=True, meanprops={"marker": "+", 
            "markeredgecolor" : "b"}, linewidth=3)
plt.show()
```

```python
sns.boxplot(x= "total_bill", y="day", data=var, hue="sex", palette='plasma', 
            order=["Fri", "Sun", "Thur", "Sat"], showmeans=True, meanprops={"marker": "+", 
            "markeredgecolor" : "b"}, linewidth=3)
plt.show()
```
