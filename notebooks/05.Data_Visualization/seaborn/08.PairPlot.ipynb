### Pair Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = sns.load_dataset("tips")
var.head(5)
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var, hue="sex")
plt.show()
```

```python
sns.pairplot(var, vars=["total_bill", "tip"] ,hue="sex")
plt.show()
```

```python
sns.pairplot(var, vars=["total_bill", "tip"] ,hue="sex", hue_order=["Female", "Male"])
plt.show()
```

```python
sns.pairplot(var, vars=["total_bill", "tip"] ,hue="sex", hue_order=["Female", "Male"], palette='BuGn')
plt.show()
```

```python
sns.pairplot(var, hue="sex", hue_order=["Female", "Male"], x_vars=["total_bill", "tip"])
plt.show()
```

```python
sns.pairplot(var, hue="sex", hue_order=["Female", "Male"], kind="reg") # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```

```python
sns.pairplot(var, hue="sex", hue_order=["Female", "Male"], kind="kde") # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```

```python
sns.pairplot(var, hue="sex", hue_order=["Female", "Male"], kind="kde", diag_kind="kde") # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```

```python
sns.pairplot(var, hue="sex", hue_order=["Female", "Male"], kind="kde", diag_kind="hist") # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```

```python
sns.pairplot(var, hue="sex", hue_order=["Female", "Male"], markers=["*", "<"]) # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```
