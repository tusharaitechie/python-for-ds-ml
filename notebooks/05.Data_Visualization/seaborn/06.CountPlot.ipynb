### Count Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset('tips')
var.head(5)
```

```python
sns.countplot(x="sex", data=var)
plt.show()
```

```python
sns.barplot(x="sex", y="size", data=var)
plt.show()
```

```python
sns.countplot(x="sex", data=var, hue="smoker")
plt.show()
```

```python
sns.countplot(y="sex", data=var, hue="smoker")
plt.show()
```

```python
sns.countplot(x="sex", data=var, hue="smoker", palette="bwr")
plt.show()
```

```python
sns.countplot(x="sex", data=var, hue="smoker", palette='dark:y')
plt.show()
```

```python
sns.countplot(x="sex", data=var, hue="smoker", palette='dark:y', saturation =0.5)
plt.show()
```
