```python
import matplotlib.pyplot as plt
day = [ 1, 2, 3, 4, 5, 6, 7 ]
no = [ 2, 3,1,4,5,3, 6 ]
colors = ["r", "y", "g", "b", "r", "g", "r"]
sizes=[400, 200, 400, 300, 200, 100, 600]
plt.scatter(day, no, c=no, s=sizes, cmap="viridis", alpha=0.5) # cmap means color maps , you can check/refer on internet
t=plt.colorbar()
t.set_label("ColorBar", fontsize=15)
plt.title("Scatter Plot", fontsize=15) #title
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.show()
```

### Two Graphs together

```python
import matplotlib.pyplot as plt
day = [ 1, 2, 3, 4, 5, 6, 7 ]
no = [ 2, 3,1,4,5,3, 6 ]
no2 = [ 3, 2, 4, 5, 1, 6, 2]
colors = [10, 49, 30, 29, 56, 20, 30]
sizes=[400, 200, 400, 300, 200, 100, 600]
plt.scatter(day, no, c=no, s=sizes, cmap="viridis", alpha=0.5)
plt.scatter(day, no2, c="r", s=sizes, alpha=0.5)
t=plt.colorbar()
t.set_label("ColorBar", fontsize=15)
plt.title("Scatter Plot", fontsize=15)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)
plt.show()
```
