### Matplotlib Box Plot

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, notch=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, vert=False)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, widths=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, tick_labels=["python"], patch_artist=True) # in new version labels replace with tick_label
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, tick_labels=["python"], patch_artist=True, showmeans=True) # in new version labels replace with tick_label
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, tick_labels=["python"], showmeans=True) # in new version labels replace with tick_label
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70, 120, 200]

plt.boxplot(x, tick_labels=["python"], showmeans=True) # in new version labels replace with tick_label
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70, 120, 200]

plt.boxplot(x, tick_labels=["python"], showmeans=True, whis=3.5) # in new version labels replace with tick_label
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70, 120, 200]

plt.boxplot(x, tick_labels=["python"], showmeans=True, sym="g+") # in new version labels replace with tick_label
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70, 120, 200]

plt.boxplot(x, tick_labels=["python"], showmeans=True, sym="g+", boxprops=dict(color="r"))
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, tick_labels=["python"], showmeans=True, sym="g+", boxprops=dict(color="r"), capprops=dict(color="r") ,whiskerprops=dict(color="r"))
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50, 60, 70]
x1 = [10, 20, 40, 50, 30, 60, 90]

y = [x, x1]
plt.boxplot(y, tick_labels=["python", "c++"], showmeans=True, sym="g+", boxprops=dict(color="r"), capprops=dict(color="r") ,whiskerprops=dict(color="r"))
plt.show()
```
