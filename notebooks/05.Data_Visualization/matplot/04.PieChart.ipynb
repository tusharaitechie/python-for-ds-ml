### Pie Charts

```python
import matplotlib.pyplot as plt 

x = [10, 20, 30, 40]
y=["c", "c++", "Java", "Python"]

plt.pie(x)
plt.show()
```

```python
import matplotlib.pyplot as plt 

x = [10, 20, 30, 40]
x1= [40, 30, 20, 10]
y=["c", "c++", "Java", "Python"]

c=["r", "b", "g", "y"]

plt.pie(x, labels=y, radius=1.5)
plt.pie([1], colors="w")

plt.show()
```

```python
import matplotlib.pyplot as plt 

x = [10, 20, 30, 40]
x1= [40, 30, 20, 10]
y=["c", "c++", "Java", "Python"]

c=["r", "b", "g", "y"]

plt.pie(x, labels=y, radius=1.5)
cr=plt.Circle(xy=(0,0), radius=1, facecolor="w")
plt.gca().add_artist(cr)

plt.show()
```
