### Outlier

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

dataset = pd.read_csv("../../datasets/loan_dataset.csv", encoding='latin1')

dataset.head(5)
```

```python
dataset.describe()
```

```python
sns.boxplot(x = "Coapplication", data = dataset)
plt.show()
```

```python
dataset.shape
```

```python
q1 = dataset["Coapplication"].quantile(0.25)
q3 = dataset["Coapplication"].quantile(0.75)
```

```python
IQR = q3-q1
```

```python
min_range = q1 - (1.5*IQR)
max_range = q3 + (1.5*IQR)
min_range, max_range
```

```python
new_dataset = dataset[dataset["Coapplication"]<=max_range]
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="Coapplication", data = new_dataset)
plt.show()
```
