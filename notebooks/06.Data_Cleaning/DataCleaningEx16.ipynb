### Handing Duplicate Values


---

# 1️⃣ Duplicate Data kya hota hai

**Duplicate rows** matlab dataset me **exactly same record 2 ya 2 se zyada baar aa gaya ho**.

Example:

| Name   | Age | Salary |             |
| ------ | --- | ------ | ----------- |
| Tushar | 25  | 50000  |             |
| Rahul  | 30  | 60000  |             |
| Tushar | 25  | 50000  | ← Duplicate |

---

# 2️⃣ Duplicate Data ke Problems

1. **Biased analysis:**
   Example: Mean salary overestimated.

2. **ML model issues:**

   * Models overfit kar sakte hain
   * Evaluation metrics inaccurate ho sakte hain

3. **Storage & performance:**
   Large dataset me unnecessary rows ka space waste

---

# 3️⃣ Detecting Duplicates in Python (pandas)

```python
import pandas as pd

data = {
    "Name":["Tushar","Rahul","Tushar"],
    "Age":[25,30,25],
    "Salary":[50000,60000,50000]
}

df = pd.DataFrame(data)

# Check duplicates
duplicates = df.duplicated()
print(duplicates)
```

Output:

```
0    False
1    False
2     True
dtype: bool
```

* `False` → Unique row
* `True` → Duplicate row

---

# 4️⃣ Count of Duplicate Rows

```python
print(df.duplicated().sum())
```

Output:

```
1
```

Ek duplicate row hai.

---

# 5️⃣ Remove Duplicate Rows

```python
df_clean = df.drop_duplicates()
print(df_clean)
```

Output:

| Name   | Age | Salary |
| ------ | --- | ------ |
| Tushar | 25  | 50000  |
| Rahul  | 30  | 60000  |

---

# 6️⃣ Remove Duplicates Based on Specific Columns

Example: Agar same `Name` aur `Age` repeat hua hai, ignore `Salary`.

```python
df_clean = df.drop_duplicates(subset=["Name","Age"])
```

---

# 7️⃣ Keep Parameter

```python
df.drop_duplicates(keep='first')  # default, first occurrence kept
df.drop_duplicates(keep='last')   # last occurrence kept
df.drop_duplicates(keep=False)    # remove all duplicates
```

---

# 8️⃣ Reset Index After Dropping

```python
df_clean = df_clean.reset_index(drop=True)
```

Ye **old index remove karke new index** assign karta hai.

---

# 9️⃣ Duplicate Detection Using SQL-Like Operations

```python
duplicates = df[df.duplicated(subset=["Name","Age"], keep=False)]
print(duplicates)
```

Ye **sab duplicates show karega**.

# 🔟 ML/DS Relevance

* **Before training ML models:**
  Duplicate data → Overfitting risk
* **Exploratory Data Analysis (EDA):**
  Statistics inaccurate ho jate hain
* **Data cleaning pipelines:**
  `drop_duplicates()` commonly used

# 11️⃣ Edge Cases

1. **Floating point differences:**

   ```python
   df.duplicated()  # may fail if 0.1 ≠ 0.1000000001
   ```

   → Round numbers first:

   ```python
   df['col'] = df['col'].round(2)
   ```

2. **Categorical differences (capital letters):**

   * `"Male"` vs `"male"` → consider same
   * Convert to lower:

   ```python
   df['Gender'] = df['Gender'].str.lower()
   ```

3. **Partial duplicates:**

   * Some columns same, others different
   * Decide `subset` carefully

---

# 12️⃣ Complete Example

```python
import pandas as pd

data = {
    "Name":["Tushar","Rahul","Tushar","Ankit","Rahul"],
    "Age":[25,30,25,28,30],
    "Salary":[50000,60000,50000,55000,60000]
}

df = pd.DataFrame(data)

print("Original Data:\n", df)

# Detect duplicates
print("Duplicate rows:\n", df[df.duplicated()])

# Drop duplicates
df_clean = df.drop_duplicates()
df_clean = df_clean.reset_index(drop=True)
print("After removing duplicates:\n", df_clean)

# Drop duplicates based on Name+Age
df_clean2 = df.drop_duplicates(subset=["Name","Age"], keep='first')
print("After removing duplicates based on Name+Age:\n", df_clean2)
```

---

# 13️⃣ Summary / Checklist

* `df.duplicated()` → Detect duplicates
* `df.drop_duplicates()` → Remove duplicates
* `subset` → Specify columns for partial duplicates
* `keep` → Decide which occurrence to keep
* `reset_index()` → Clean index
* Always clean before ML / EDA

---

💡 **Pro Tip:**

* Duplicate handling **first step of data cleaning**
* Combine with **missing value handling** → Clean dataset ready for scaling, encoding, ML.

# 1️⃣ Method 1: Inplace Update

Pandas me kaafi functions me **`inplace=True`** option hota hai. Ye directly **original DataFrame ko update kar deta hai**.

```python
import pandas as pd

# Sample dataset
data = {
    "Name":["Tushar","Rahul","Tushar","Ankit","Rahul"],
    "Age":[25,30,25,28,30],
    "Salary":[50000,60000,50000,55000,60000]
}

df = pd.DataFrame(data)

print("Original df:\n", df)

# Drop duplicates directly in original df
df.drop_duplicates(inplace=True)

print("Updated df:\n", df)
```

Output:

| Name   | Age | Salary |
| ------ | --- | ------ |
| Tushar | 25  | 50000  |
| Rahul  | 30  | 60000  |
| Ankit  | 28  | 55000  |

✅ **Original `df` ab updated ho gaya**, koi nayi variable banayi nahi.

---

# 2️⃣ Method 2: Overwrite Original Dataset

Agar function **inplace=False** default hai, to naya DataFrame return hota hai. Tum **original variable me overwrite** kar sakte ho:

```python
df = df.drop_duplicates()
```

* Yaha nayi DataFrame assign ho gayi **original variable `df` me**
* Result same as `inplace=True`

---

# 3️⃣ Subset ke saath Update

Agar tum **specific columns** ke basis pe duplicate remove karna chahte ho:

```python
# Remove duplicates based on Name and Age, update original df
df.drop_duplicates(subset=["Name","Age"], keep='first', inplace=True)
```

* First occurrence kept
* Original dataset me reflect ho gaya

---

# 4️⃣ Reset Index (Optional but recommended)

Drop duplicates ke baad **index mismatch ho sakta hai**. Original DataFrame me update karne ke liye:

```python
df.reset_index(drop=True, inplace=True)
```

* `drop=True` → old index remove
* Index **0,1,2,…** se start hota hai

---

# 5️⃣ Summary Table

| Operation                       | Effect on original DataFrame     |
| ------------------------------- | -------------------------------- |
| `drop_duplicates()`             | Returns new DataFrame (default)  |
| `drop_duplicates(inplace=True)` | Updates original DataFrame       |
| `df = df.drop_duplicates()`     | Original variable overwritten    |
| `reset_index(inplace=True)`     | Original DataFrame index updated |

---

# ✅ Example: Full Flow

```python
import pandas as pd

df = pd.DataFrame({
    "Name":["Tushar","Rahul","Tushar","Ankit","Rahul"],
    "Age":[25,30,25,28,30],
    "Salary":[50000,60000,50000,55000,60000]
})

print("Original Data:\n", df)

# Remove duplicates based on Name+Age
df.drop_duplicates(subset=["Name","Age"], keep='first', inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)

print("Updated Original DataFrame:\n", df)
```

Output:

| Name   | Age | Salary |
| ------ | --- | ------ |
| Tushar | 25  | 50000  |
| Rahul  | 30  | 60000  |
| Ankit  | 28  | 55000  |

---

💡 **Key Point:**

* `inplace=True` → direct update
* Nahi to variable ko **overwrite karna padega**
* Index reset karna optional hai, readability ke liye best practice
