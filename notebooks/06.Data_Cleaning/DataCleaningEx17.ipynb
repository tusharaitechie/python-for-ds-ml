### Replace Value and change the DataType

```python
import pandas as pd

dataset = pd.read_csv("../../datasets/loan_dataset.csv", encoding='latin1')

dataset.head(5)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset["Dependend"].value_counts()
```

```python
dataset["Dependend"].fillna(dataset["Dependend"].mode()[0])
```

```python
dataset["Dependend"].replace("3+", "3", inplace=True)
```

```python
# dataset["Dependend"]=dataset["Dependend"].astype("int64")
# converts datatype using astype function
dataset["Dependend"] = pd.to_numeric(dataset["Dependend"], errors="coerce") 
```
