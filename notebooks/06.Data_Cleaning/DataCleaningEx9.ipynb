### Label Encoding

```python
import pandas as pd

df = pd.DataFrame({"name": ["wscube", "cow", "cat", "dog", "black"]})
df
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
le = LabelEncoder()
df["en_name"] = le.fit_transform(df["name"])
```

```python
df
```

```python
dataset = pd.read_csv("../../datasets/loan_dataset.csv", encoding='latin1')

dataset.head(5)
```

```python
dataset["Property_Area"].unique()
```

```python
la = LabelEncoder()
la.fit(dataset["Property_Area"])
```

```python
la.transform(dataset["Property_Area"])
```

```python
dataset["Property_Area"] = la.transform(dataset["Property_Area"])
```
