### Ordinal Encoding

 we can perform ordinal encoding by two ways 1. Cyclic Line 2. Map Function

#### Cyclic Line Method

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Size": ["s", "m", "l", "xl", "s", "m", "l", "s", "s", "l", "sl", "m"]})
df.head(5)
```

```python
ord_data = [["s", "m", "l", "xl", "sl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data) 
# oe = OrdinalEncoder() its called Automatic Categories detection
oe.fit(df[["Size"]])
```

```python
oe.transform(df[["Size"]])
```

```python
df["Size_en"] = oe.transform(df[["Size"]])
```

```python
df
```

#### Map Function Method

```python
ord_data1 = {"s": 0, "m":1, "l":2, "x1":3}
```

```python
df["Size"].map(ord_data1)
```

```python
df["Size_en_map"] = df["Size"].map(ord_data1)
```

```python
df
```
