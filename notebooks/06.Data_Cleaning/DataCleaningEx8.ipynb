```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("../../datasets/loan_dataset.csv", encoding='latin1')

dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset["Gender"] = dataset["Gender"].fillna(dataset["Gender"].mode()[0])
```

```python
dataset["Married"] = dataset["Married"].fillna(dataset["Married"].mode()[0])
```

```python
en_data = dataset[["Gender", "Married"]]
```

### Method 1

```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

### Method 2

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ar = ohe.fit_transform(en_data).toarray()
ohe.get_feature_names_out()
```

```python
pd.DataFrame(ar, columns=["Gender_Female", "Gender_Male", "Married_No", "Married_Yes"])
```

```python
ohe = OneHotEncoder(drop="first") # by using drop first metho first columsn are going to be deleted in our case Gender_Female & Married_No will be delted 
ar = ohe.fit_transform(en_data).toarray() #convert categorical data into numberical data
ar
```

```python
ohe.get_feature_names_out()
```

```python
pd.DataFrame(ar, columns=["Gender_Male", "Married_Yes"])
```
