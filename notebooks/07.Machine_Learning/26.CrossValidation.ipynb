```python
import pandas as pd
from sklearn.model_selection import KFold, cross_val_score

dataset = pd.read_csv("../../datasets/placement.csv")

dataset.head(5)
```

```python
x = dataset.iloc[:, :-1]
y = dataset["package"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import cross_val_score
```

```python
p3 = cross_val_score(LinearRegression(), x, y, cv=5)
```

```python
p3.sort()
p3
```

```python
p3.sort()
p3*100
```

```python
p4 = cross_val_score(LinearRegression(), x, y, cv=KFold(n_splits=10))
```

```python
p4.sort()
p4
```

```python
p4.sort()
p4*100
```

```python
new_data = dataset.head(10)
```

```python
x_new = new_data.iloc[:, :-1]
y_new = new_data["package"]
```

```python
from sklearn.model_selection import LeaveOneOut, LeavePOut, KFold, StratifiedKFold
```

```python
lo = LeaveOneOut()

for train, test in lo.split(x, y):
    print(train, test)
```

```python
lo1 = LeavePOut(p=2)

for train, test in lo1.split(x_new, y_new):
    print(train, test)
```

```python
lo2 = KFold(n_splits=5)

for train, test in lo1.split(x_new, y_new):
    print(train, test)
```
