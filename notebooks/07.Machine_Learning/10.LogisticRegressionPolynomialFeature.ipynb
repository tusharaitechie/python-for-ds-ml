### Logistic Regression (Binary Classification) Polynomial Feature

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_excel("../../datasets/polynomical classication.xlsx")

dataset.head(5)
```

```python
sns.scatterplot(x="Data1", y="Data2", data=dataset, hue="Output")
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset["Output"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2) # change the degree for accuracy
pf.fit(x)
x = pd.DataFrame(pf.transform(x))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```
