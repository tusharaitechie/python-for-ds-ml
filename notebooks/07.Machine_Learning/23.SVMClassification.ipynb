```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

dataset = pd.read_excel("../../datasets/placement.xlsx")

dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="CGPA", y="Score", data=dataset, hue="Placed")
plt.show()
```

```python
x=dataset.iloc[:, : -1]
y = dataset["Placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
sv = SVC(kernel="linear") # different types of kernals : 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)*100
```

```python
sv.score(x_train, y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=sv)
plt.show()
```
