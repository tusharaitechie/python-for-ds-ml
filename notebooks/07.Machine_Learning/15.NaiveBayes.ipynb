### Naive Bayes

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

dataset = pd.read_excel("../../datasets/placement.xlsx")
dataset.head(5)
```

```python
sns.kdeplot(data=dataset["CGPA"])
plt.show()
```

```python
sns.kdeplot(data=dataset["Score"])
plt.show()
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="CGPA", y="Score", data=dataset, hue="Placed")
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset["Placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
gnb.score(x_test, y_test)*100, gnb.score(x_train, y_train)*100
```

```python
gnb.predict([[7.5,70]])
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=gnb)
plt.show()
```

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
mnb.score(x_train, y_train)*100, mnb.score(x_test, y_test)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=mnb)
plt.show()
```

```python
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
bnb.score(x_train, y_train)*100, bnb.score(x_test, y_test)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=bnb)
plt.show()
```
