**Decision Tree** में **Pruning (छँटाई)** का मतलब होता है पेड़ को छोटा और simple बनाना ताकि **overfitting** कम हो जाए।

Pruning के दो प्रकार होते हैं 👇

---

## 🌳 1. Pre-Pruning (प्री-प्रूनिंग)

👉 इसे **Early Stopping** भी कहते हैं

### 🔹 क्या होता है?

Tree को **पूरी तरह बनने से पहले ही रोक दिया जाता है**

👉 यानी:

* अगर आगे split करने से फायदा नहीं है → वहीं रुक जाते हैं

---

### 🔹 कैसे रोकते हैं?

कुछ conditions लगाई जाती हैं:

* max depth (पेड़ की अधिकतम ऊँचाई)
* minimum samples (कम से कम data होना चाहिए)
* minimum **Information Gain** (अगर gain कम है → split नहीं)

---

### 🔹 फायदे:

✔️ Fast (जल्दी बनता है)
✔️ Overfitting कम करता है

### 🔹 नुकसान:

❌ कभी-कभी जल्दी रुक जाता है → underfitting हो सकता है

---

## 🌳 2. Post-Pruning (पोस्ट-प्रूनिंग)

👉 इसे **Backward Pruning** भी कहते हैं

### 🔹 क्या होता है?

* पहले पूरा tree बनाते हैं
* फिर धीरे-धीरे बेकार branches काटते हैं ✂️

---

### 🔹 कैसे करते हैं?

* उन branches को हटाते हैं जो model की accuracy improve नहीं कर रहीं
* validation data का उपयोग करते हैं

---

### 🔹 फायदे:

✔️ ज्यादा accurate होता है
✔️ Better generalization

### 🔹 नुकसान:

❌ Time ज्यादा लगता है
❌ Computationally expensive

---

## 🔹 आसान तुलना:

| Feature    | Pre-Pruning      | Post-Pruning     |
| ---------- | ---------------- | ---------------- |
| कब होता है | Tree बनाते समय   | Tree बनने के बाद |
| Speed      | तेज              | धीमा             |
| Risk       | Underfitting     | कम risk          |
| Accuracy   | थोड़ी कम हो सकती | ज्यादा अच्छी     |

---

## 🔹 एक लाइन में समझो:

* **Pre-Pruning** = “पहले ही रोक दो”
* **Post-Pruning** = “पहले बनाओ, फिर काटो”
